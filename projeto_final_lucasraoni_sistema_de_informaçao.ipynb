```python
#1. Projeto Desenvolvido — Sistema CRUD de Usuários
# Lista para armazenar os usuários
usuarios = []

def mostrar_menu():
    print("\n--- MENU ---")
    print("1 - Cadastrar usuário")
    print("2 - Listar usuários")
    print("3 - Atualizar usuário")
    print("4 - Excluir usuário")
    print("5 - Sair")

while True:
    mostrar_menu()
    opcao = int(input("Escolha uma opção: "))

    # Cadastrar
    if opcao == 1:
        nome = input("Digite o nome: ")
        idade = int(input("Digite a idade: "))
        usuarios.append({"nome": nome, "idade": idade})
        print("Usuário cadastrado com sucesso!")

    # Listar
    elif opcao == 2:
        if len(usuarios) == 0:
            print("Nenhum usuário cadastrado.")
        else:
            print("\n--- Lista de Usuários ---")
            for i, u in enumerate(usuarios):
                print(f"{i} - Nome: {u['nome']}, Idade: {u['idade']}")

    # Atualizar
    elif opcao == 3:
        indice = int(input("Digite o índice do usuário para atualizar: "))
        if 0 <= indice < len(usuarios):
            novo_nome = input("Novo nome: ")
            nova_idade = int(input("Nova idade: "))
            usuarios[indice]["nome"] = novo_nome
            usuarios[indice]["idade"] = nova_idade
            print("Usuário atualizado com sucesso!")
        else:
            print("Índice inválido.")

    # Excluir
    elif opcao == 4:
        indice = int(input("Digite o índice do usuário para excluir: "))
        if 0 <= indice < len(usuarios):
            usuarios.pop(indice)
            print("Usuário removido com sucesso!")
        else:
            print("Índice inválido.")

    # Sair
    elif opcao == 5:
        print("Encerrando o sistema...")
        break

    else:
        print("Opção inválida, tente novamente.")
```

```python
#2. Fluxograma do Sistema CRUD
  ┌───────────┐
  │   Início   │
  └─────┬─────┘
        v
  ┌───────────┐
  │  Menu      │
  └─────┬─────┘
        v
  ┌─────────────────────────┐
  │ Opção 1? Cadastrar       │─── Sim → Cadastra
  └─────┬───────────────────┘
        │ Não
        v
  ┌─────────────────────────┐
  │ Opção 2? Listar          │─── Sim → Lista
  └─────┬───────────────────┘
        │ Não
        v
  ┌─────────────────────────┐
  │ Opção 3? Atualizar       │─── Sim → Atualiza
  └─────┬───────────────────┘
        │ Não
        v
  ┌─────────────────────────┐
  │ Opção 4? Excluir         │─── Sim → Remove
  └─────┬───────────────────┘
        │ Não
        v
  ┌─────────────────────────┐
  │ Opção 5? Sair            │─── Sim → FIM
  └─────┬───────────────────┘
        │ Não
        v
  Volta ao Menu
```

```python
#3. Linguagem Algorítmica
# Algoritmo CRUD_Usuarios
# Declaração das variáveis/estrutura
usuarios = []   # vetor/lista para armazenar usuários
opcao = 0       # controle do menu
# Início do algoritmo
while opcao != 5:
    print("1 - Cadastrar")
    print("2 - Listar")
    print("3 - Atualizar")
    print("4 - Excluir")
    print("5 - Sair")
    opcao = int(input("Escolha uma opção: "))
    # Opção 1 - Cadastrar
    if opcao == 1:
        nome = input("Digite o nome: ")
        idade = int(input("Digite a idade: "))
        usuarios.append({"nome": nome, "idade": idade})
    # Opção 2 - Listar
    elif opcao == 2:
        if len(usuarios) == 0:
            print("Nenhum usuário cadastrado")
        else:
            for i, u in enumerate(usuarios):
                print(i, "-", u["nome"], "-", u["idade"])
    # Opção 3 - Atualizar
    elif opcao == 3:
        indice = int(input("Digite o índice para atualizar: "))
        if 0 <= indice < len(usuarios):
            nome = input("Novo nome: ")
            idade = int(input("Nova idade: "))
            usuarios[indice] = {"nome": nome, "idade": idade}
        else:
            print("Índice inválido")
    # Opção 4 - Excluir
    elif opcao == 4:
        indice = int(input("Digite o índice para excluir: "))
        if 0 <= indice < len(usuarios):
            usuarios.pop(indice)
        else:
            print("Índice inválido")
    # Opção 5 - Sair
    elif opcao == 5:
        print("Encerrando o algoritmo...")
    else:
        print("Opção inválida")
```

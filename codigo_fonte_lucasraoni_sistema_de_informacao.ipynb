```python
# ------------------------------------------------------------
#  PROJETO CRUD
#  Sistema simples para cadastrar, listar, atualizar e excluir
#  usuários usando lista e dicionários.
# ------------------------------------------------------------

# Lista onde os usuários serão armazenados
usuarios = []
# -------- Função para exibir o menu --------
def mostrar_menu():
    print("\n====== MENU PRINCIPAL ======")
    print("1 - Cadastrar usuário")
    print("2 - Listar usuários")
    print("3 - Atualizar usuário")
    print("4 - Excluir usuário")
    print("5 - Sair")
    print("============================")
# -------- Função para cadastrar usuário --------

def cadastrar_usuario():
    nome = input("Digite o nome do usuário: ")
    idade = int(input("Digite a idade do usuário: "))
    usuarios.append({"nome": nome, "idade": idade})
    print("Usuário cadastrado com sucesso!")

# -------- Função para listar usuários --------
def listar_usuarios():
    if len(usuarios) == 0:
        print("Nenhum usuário cadastrado.")
    else:
        print("\n------ Lista de Usuários ------")
        for i, u in enumerate(usuarios):
            print(f"{i} - Nome: {u['nome']} | Idade: {u['idade']}")
        print("-------------------------------")


# -------- Função para atualizar usuário --------
def atualizar_usuario():
    listar_usuarios()
    indice = int(input("\nDigite o índice do usuário a atualizar: "))

    if 0 <= indice < len(usuarios):
        novo_nome = input("Novo nome: ")
        nova_idade = int(input("Nova idade: "))
        usuarios[indice]["nome"] = novo_nome
        usuarios[indice]["idade"] = nova_idade
        print("Usuário atualizado com sucesso!")
    else:
        print("Índice inválido!")


# -------- Função para excluir usuário --------
def excluir_usuario():
    listar_usuarios()
    indice = int(input("\nDigite o índice do usuário a excluir: "))

    if 0 <= indice < len(usuarios):
        removido = usuarios.pop(indice)
        print(f"Usuário '{removido['nome']}' removido com sucesso!")
    else:
        print("Índice inválido!")

# -------- Função principal --------
def main():
    while True:
        mostrar_menu()
        opcao = int(input("Escolha uma opção: "))

        if opcao == 1:
            cadastrar_usuario()
        elif opcao == 2:
            listar_usuarios()
        elif opcao == 3:
            atualizar_usuario()
        elif opcao == 4:
            excluir_usuario()
        elif opcao == 5:
            print("Encerrando o sistema... Até logo!")
            break
        else:
            print("Opção inválida. Tente novamente.")

# -------- Execução do programa --------
if __name__ == "__main__":
    main()
```
